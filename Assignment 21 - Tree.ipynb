```python
class newNode:
    def __init__(self, data):
        self.data = data
        self.left = self.right = None
 
 
""" Change a tree so that the roles of the
    left and right pointers are swapped at
    every node.
 
So the tree...
        4
        / \
    2 5
    / \
    1 3
 
is changed to...
    4
    / \
    5 2
    / \
    3 1
"""
 
 
def mirror(node):
 
    if (node == None):
        return
    else:
 
        temp = node
 
        """ do the subtrees """
        mirror(node.left)
        mirror(node.right)
 
        """ swap the pointers in this node """
        temp = node.left
        node.left = node.right
        node.right = temp
 
 
""" Helper function to print Inorder traversal."""
 
 
def inOrder(node):
 
    if (node == None):
        return
 
    inOrder(node.left)
    print(node.data, end=" ")
    inOrder(node.right)
 
 
# Driver code
if __name__ == "__main__":
 
    root = newNode(1)
    root.left = newNode(2)
    root.right = newNode(3)
    root.left.left = newNode(4)
    root.left.right = newNode(5)
 
    """ Print inorder traversal of
        the input tree """
    print("Inorder traversal of the",
          "constructed tree is")
    inOrder(root)
 
    """ Convert tree to its mirror """
    mirror(root)
 
    """ Print inorder traversal of
        the mirror tree """
    print("\nInorder traversal of",
          "the mirror tree is ")
    inOrder(root)
```

```python
from collections import deque

# Node class definition
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.next = None

# Function to connect nodes at the same level
def connect_nodes(root):
    if not root:
        return

    # Create a queue for BFS
    queue = deque()
    queue.append(root)

    while queue:
        level_size = len(queue)
        prev = None

        # Process nodes at the current level
        for _ in range(level_size):
            node = queue.popleft()

            # Connect to the previous node
            if prev:
                prev.next = node

            prev = node

            # Add the child nodes to the queue
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        # Connect the last node in the level to None
        prev.next = None

# Function to print the connections
def print_connections(root):
    node = root
    while node:
        print(node.data, end=" â†’ ")
        if node.next:
            print(node.next.data)
        else:
            print("-1")
        node = node.next

# Create the tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connect_nodes(root)

# Print the connections
print("Output:")
print_connections(root)
```
